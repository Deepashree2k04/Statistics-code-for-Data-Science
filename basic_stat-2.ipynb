```python
import numpy as np
import pandas as pd
```

1.	Load the patient dataset and show summary with info().

```python
patient_data = pd.read_csv("Patient_Data.csv")
billing_data = pd.read_csv("Billing_Data.csv")
```

```python
patient_data
```

```python
patient_data.info()
```

2.	Select only the columns relevant for billing: ['PatientID', 'Department', 'Doctor', 'BillAmount'].

```python
patient_data[['PatientID', 'Department', 'Doctor', 'BillAmount']]
```

3.	Drop administrative columns like ['ReceptionistID', 'CheckInTime'].

```python
patient_data = patient_data.drop(['ReceptionistID', 'CheckInTime'], axis=1)
patient_data
```

4.	Use groupby to find total bill amount per department.

```python
patient_data.groupby('Department')["BillAmount"].sum()
```

5.	Remove duplicate patient records based on PatientID.

```python
patient_data = patient_data.drop_duplicates('PatientID')
patient_data
```

6.	Fill missing BillAmount values with the mean bill amount.

```python
patient_data = patient_data.fillna(patient_data['BillAmount'].mean())
```

```python
patient_data
```

7.	Merge the billing dataset with patient dataset on PatientID.

```python
merged_data = pd.merge(patient_data, billing_data, on='PatientID')
```

```python
merged_data
```

8.	Concatenate an additional DataFrame that contains new patients for the current week (row-wise).

```python
new_patients_df = pd.DataFrame({
    'PatientID': [105, 106, 107],
    'Name': ['Meera', 'Kiran', 'Joseph'],
    'Department': ['Cardiology', 'Orthopedics', 'Dermatology'],
    'Doctor' : ['Dr. John', 'Dr. Smith', 'Dr. Lee'],
    'BillAmount': [4500, 3200, 2900]
})
```

```python
patient_data = pd.concat([patient_data, new_patients_df], axis=0, ignore_index=True)
```

```python
patient_data
```

9.	Concatenate new billing category columns like ['InsuranceCovered', 'FinalAmount'] (column-wise)

```python
new_billing = pd.DataFrame({
'InsuranceCovered' : [2000,3000,3500,2500,3000,1000,500,0],
'FinalAmount' : [3000,3233.333333,4000,3700,3233.333333,3500,2700,2900]
}
)
```

```python
patient_data = pd.concat([patient_data,new_billing], axis=1,)
```
